```python
import matplotlib.pyplot as plt
import numpy as np


xc = np.linspace(-100, 100, 100000)
yc = xc*np.sin((xc*xc)/300)

plt.plot(xc, yc)
plt.show()


```

#### Part 2 build models

In the cells below, build your models and train it with data from part1.
<br>You need to split your training data to two parts. With 40% used in training, and 60% used in test.

```python
# model 1
from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Activation, Flatten
from sklearn.model_selection import train_test_split

#X_train = xc
#Y_train = yc

X_train, X_test, Y_train, Y_test = train_test_split(xc, yc, test_size=0.6, train_size=0.4, random_state=42)

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(30,activation = "relu"))
model.add(Dense(30,activation = "relu"))
model.add(Dense(30,activation = "relu"))
model.add(Dense(30,activation = "relu"))
model.add(Dense(30,activation = "relu"))
model.add(Dense(1))

model.compile(
   loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)
model.fit(X_train, Y_train, batch_size = 128, epochs = 100)

```
